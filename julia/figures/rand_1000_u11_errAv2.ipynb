```julia
include("../src/Julia.jl")
```

```julia
prec = ["default", "64", "128", "256", "512", "1024"];
```

```julia
jul = []
for p in 1:6
    push!(jul, parseData("../../graphs/pathRand_1000_u11/_log_julia_tree1_" * prec[p] * ".txt"))
end
```

```julia
#2 norm plot
datajA = [Array{Float64,1}(0) for i in 1:6]

for i in 1:100
    for p in 1:6
        push!(datajA[p], jul[p][i][1])
    end
end
```

```julia
for p in 1:6
    plot(log(datajA[p]), label="Julia " * prec[p], linewidth=2)
end

legend(bbox_to_anchor=(1.05, 1), loc=2, fontsize=20, borderaxespad=0.)

ylabel("Relative A-norm")
xlabel("PCG Iterations")
```
