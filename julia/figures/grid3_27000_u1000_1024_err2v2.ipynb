```julia
jul = []
for t in 1:5
    push!(jul, parseData("../../graphs/grid3_27000_u1000/_log_julia_tree$(t)_1024.txt"))
end
```

```julia
#2 norm plot
dataj2 = [Array{Float64,1}(0) for i in 1:5]

for i in 1:600
    for t in 1:5
        push!(dataj2[t], jul[t][i][2])
    end
end
```

```julia
name = ["AKPW", "Greedy Prim", "Greedy Kruskal", "MST", "Shortest Path"]
for t in 1:5
    plot(log(dataj2[t]), label=name[t], linewidth=2)
end

ax = gca()
ax[:set_ylim]([-40,0.25])

# legend(bbox_to_anchor=(1.05, 1), loc=2, fontsize=15, borderaxespad=0.)
ylabel("Relative 2-norm")
xlabel("PCG Iterations")
```
