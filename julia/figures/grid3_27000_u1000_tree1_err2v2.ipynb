```julia
include("../src/Julia.jl")
```

```julia
prec = ["default", "64", "128", "256", "512", "1024"];
```

```julia
jul = []
for p in 1:6
    push!(jul, parseData("../../graphs/grid3_27000_u1000/_log_julia_tree1_" * prec[p] * ".txt"))
end
```

```julia
#2 norm plot
dataj2 = [Array{Float64,1}(0) for i in 1:6]

for i in 1:300
    for p in 1:6
        push!(dataj2[p], jul[p][i][2])
    end
end
```

```julia
for p in 1:6
    plot(log(dataj2[p]), label="Julia " * prec[p], linewidth=2)
end

legend(bbox_to_anchor=(1.05, 1), loc=2, fontsize=20, borderaxespad=0.)
ylabel("Relative 2-norm")
xlabel("PCG Iterations")
```
