```julia
jul64 = parseData("../../graphs/rand_1000_u1000/_log_julia_tree1_64.txt")
pyt64 = parsePythonData("../../python/_log_python_tree_1_39_100iters.csv")
mat64 = parseMatlabData("../../matlab/_log_matlab_tree1_64precision_100iters.txt")

jul1024= parseData("../../graphs/rand_1000_u1000/_log_julia_tree1_1024.txt")
pyt1024 = parsePythonData("../../python/_log_python_tree_1_309_100iters.csv")
mat1024 = parseMatlabData("../../matlab/_log_matlab_tree1_1024precision_100iters.txt");
```

```julia
#2 norm plot
dataj2 = [Array{Float64,1}(0) for i in 1:2]
datap2 = [Array{Float64,1}(0) for i in 1:2]
datam2 = [Array{Float64,1}(0) for i in 1:2]

for i in 1:100
    push!(dataj2[1], jul64[i][2])
    push!(dataj2[2], jul1024[i][2])
    
    push!(datap2[1], pyt64[i][2])
    push!(datap2[2], pyt1024[i][2])
    
    push!(datam2[1], mat64[i][2])
    push!(datam2[2], mat1024[i][2])
end
```

```julia
plot(log(dataj2[1]), label="julia 64", linewidth=2.5)
plot(log(dataj2[2]), label="julia 1024", linewidth=2.5)

plot(log(datap2[1]), label="python 64", linewidth=2.5)
plot(log(datap2[2])-0.1, label="python 1024", linewidth=2.5)

plot(log(datam2[1]), label="matlab 64", linewidth=2.5)
plot(log(datam2[2])-0.2, label="matlab 1024", linewidth=2.5)

legend(fontsize=7)
ylabel("Relative 2-norm")
xlabel("Iterations")
```
